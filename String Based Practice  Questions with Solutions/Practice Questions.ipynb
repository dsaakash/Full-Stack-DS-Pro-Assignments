```python
# 1. Reverse a string
def reverse_string(input_str):
    return input_str[::-1]
```

```python
# 2. Check if a string is a palindrome
def is_palindrome(input_str):
    reversed_str = input_str[::-1]
    return input_str == reversed_str
```

```python
# 3. Convert a string to uppercase
def convert_to_uppercase(input_str):
    return input_str.upper()
```

```python
# 4. Convert a string to lowercase
def convert_to_lowercase(input_str):
    return input_str.lower()
```

```python
# 5. Count the number of vowels in a string
def count_vowels(input_str):
    vowels = "aeiouAEIOU"
    return sum(1 for char in input_str if char in vowels)
```

```python
# 6. Count the number of consonants in a string
def count_consonants(input_str):
    consonants = "bcdfghjklmnpqrstvwxyzBCDFGHJKLMNPQRSTVWXYZ"
    return sum(1 for char in input_str if char in consonants)
```

```python
# 7. Remove all whitespaces from a string
def remove_whitespaces(input_str):
    return "".join(input_str.split())
```

```python
# 8. Find the length of a string without using the `len()` function
def string_length(input_str):
    count = 0
    for char in input_str:
        count += 1
    return count
```

```python
# 9. Check if a string contains a specific word
def contains_word(input_str, word):
    return word in input_str
```

```python
# 10. Replace a word in a string with another word
def replace_word(input_str, old_word, new_word):
    return input_str.replace(old_word, new_word)
```

```python
# 11. Count the occurrences of a word in a string
def count_occurrences(input_str, word):
    return input_str.count(word)
```

```python
# 12. Find the first occurrence of a word in a string
def find_first_occurrence(input_str, word):
    return input_str.find(word)
```

```python
# 13. Find the last occurrence of a word in a string
def find_last_occurrence(input_str, word):
    return input_str.rfind(word)
```

```python
# 14. Split a string into a list of words
def split_string(input_str):
    return input_str.split()
```

```python
# 15. Join a list of words into a string
def join_words(word_list):
    return " ".join(word_list)

```

```python
import string
from urllib.parse import urlparse
from collections import Counter

# 31. Extract the username from an email address string
def extract_username(email_address):
    return email_address.split('@')[0]

# 32. Extract the domain name from an email address string
def extract_domain(email_address):
    return email_address.split('@')[1]

# 33. Replace multiple spaces with a single space
def replace_multiple_spaces(input_str):
    return ' '.join(input_str.split())

# 34. Check if a string is a valid URL
def is_valid_url(url):
    try:
        result = urlparse(url)
        return all([result.scheme, result.netloc])
    except ValueError:
        return False

# 35. Extract the protocol from a URL string
def extract_protocol(url):
    return urlparse(url).scheme if is_valid_url(url) else None

# 36. Find the frequency of each character in a string
def character_frequency(input_str):
    return Counter(input_str)

# 37. Remove all punctuation from a string
def remove_punctuation(input_str):
    return input_str.translate(str.maketrans("", "", string.punctuation))

# 38. Check if a string contains only digits
def contains_only_digits(input_str):
    return input_str.isdigit()

# 39. Check if a string contains only alphabets
def contains_only_alphabets(input_str):
    return input_str.isalpha()

# 40. Convert a string to a list of characters
def string_to_list(input_str):
    return list(input_str)

# 41. Check if two strings are anagrams
def are_anagrams(str1, str2):
    return Counter(str1) == Counter(str2)

# 42. Encode a string using a Caesar cipher
def caesar_cipher_encode(input_str, shift):
    result = ""
    for char in input_str:
        if char.isalpha():
            ascii_offset = ord('A') if char.isupper() else ord('a')
            result += chr((ord(char) - ascii_offset + shift) % 26 + ascii_offset)
        else:
            result += char
    return result

# 43. Decode a Caesar cipher encoded string
def caesar_cipher_decode(encoded_str, shift):
    return caesar_cipher_encode(encoded_str, -shift)

# 44. Find the most frequent word in a string
def most_frequent_word(input_str):
    words = input_str.split()
    word_counts = Counter(words)
    return word_counts.most_common(1)[0][0]

# 45. Find all unique words in a string
def unique_words(input_str):
    return list(set(input_str.split()))

# 46. Count the number of syllables in a string
def count_syllables(input_str):
    vowels = "aeiouyAEIOUY"
    count = 0
    for word in input_str.split():
        for char in word:
            if char in vowels:
                count += 1
    return count

# Example usage:
email_address = "user@example.com"
url = "https://www.example.com"
caesar_input = "Hello, World!"
word_frequency_input = "this is a sample sentence. this is another sentence."

# Test each function
print("31. Extract username:", extract_username(email_address))
print("32. Extract domain:", extract_domain(email_address))
print("33. Replace multiple spaces:", replace_multiple_spaces("   Hello   World!   "))
print("34. Is valid URL:", is_valid_url(url))
print("35. Extract protocol:", extract_protocol(url))
print("36. Character frequency:", character_frequency(word_frequency_input))
print("37. Remove punctuation:", remove_punctuation("Hello, World! This is a sample sentence."))
print("38. Contains only digits:", contains_only_digits("12345"))
print("39. Contains only alphabets:", contains_only_alphabets("HelloWorld"))
print("40. String to list:", string_to_list("Hello, World!"))
print("41. Are anagrams:", are_anagrams("listen", "silent"))
print("42. Caesar cipher encode:", caesar_cipher_encode(caesar_input, 3))
print("43. Caesar cipher decode:", caesar_cipher_decode("Khoor, Zruog!", 3))
print("44. Most frequent word:", most_frequent_word(word_frequency_input))
print("45. Unique words:", unique_words(word_frequency_input))
print("46. Count syllables:", count_syllables("Hello World! This is a sample sentence."))
```

```python
import csv
import string

# 47. Check if a string contains any special characters
def contains_special_characters(input_str):
    # Using string.punctuation to check for special characters
    return any(char in string.punctuation for char in input_str)

# 48. Remove the nth word from a string
def remove_nth_word(input_str, n):
    words = input_str.split()
    if 1 <= n <= len(words):
        del words[n - 1]
        return ' '.join(words)
    else:
        return "Invalid index. Please provide a valid index."

# 49. Insert a word at the nth position in a string
def insert_word_at_nth_position(input_str, n, word_to_insert):
    words = input_str.split()
    if 0 <= n <= len(words):
        words.insert(n, word_to_insert)
        return ' '.join(words)
    else:
        return "Invalid index. Please provide a valid index."

# 50. Convert a CSV string to a list of lists
def csv_to_list_of_lists(csv_string):
    # Using the csv module to parse the CSV string without newline parameter
    csv_list = list(csv.reader([line.strip() for line in csv_string.split('\n')]))
    return csv_list

# Example usage:
special_characters_string = "Hello, World! This is a string with @ and #."
nth_removal_string = "This is a sample sentence for nth word removal."
nth_insertion_string = "This is a sample sentence for nth word insertion."
csv_string = "Name,Age,Location\nJohn,25,New York\nJane,30,San Francisco"

# Test each function
print("47. Contains special characters:", contains_special_characters(special_characters_string))
print("48. Remove nth word:", remove_nth_word(nth_removal_string, 3))
print("49. Insert word at nth position:", insert_word_at_nth_position(nth_insertion_string, 2, "new"))
print("50. CSV string to list of lists:", csv_to_list_of_lists(csv_string))
```
